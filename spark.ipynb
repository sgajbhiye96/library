```python
import pandas as pd
pdf=pd.read_excel("")
df=spark.CreateDataFrame(pdf)
```

```python
# Data Cleaning

df.show()
from pyspark.sql.functions import col
df.withColumn("price",col("price").cast("double"))
```

```python
df.printSchema()
df.dtypes
```

```python
from pyspark.sql.functions import sum,when

df.select([sum(when(col(c).isNull(),1).otherwise(0).alias(c))
           for c in df.columns]).show()
```

```python
df=df.dropna(subset(["price","city"]))
```

```python
df=df.fillna("city":"Unknown")
```

```python
df.dropDuplicates(['order_id'])
from pyspark.sql.functions import to_date
df.withColumn("city",upper(trim((col("city"))))
df.withColumn("date",to_date(col("date")))          

```

```python
df=df.withColumn("final Price",col("price")*0.9)

from pyspark.sql.functions import month

df=df.withColumn("month",month(col("date")))

from pyspark.sql.functions import when

df=df.withColumn("Price Category",
                 when(col("price")<300,"Low").when(col("price")<800,"Medium").otherwise("High"))


df.describe().show()

df.groupby("city").count().show()

df.groupby("city").sum("price").show()

df.filter(col("price").isNull().show())

df.stat.corr("price","user_id")

df.write.mode("overwrite").parque("orders")
```
