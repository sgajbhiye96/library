```python
def smartdiv(func):
    def inner(x,y):
        if x<y:
            x,y=y,x
        return func(x,y)
    return inner
```

```python
@smartdiv
def div(x,y):
    return x/y
```

```python
import time
def timer(func):
    def wrapper(*a,**k):
        start=time.time()
        result=func(*a,**k)
        end=time.time()

        print("Time",end-start)
    return wrapper
    




```

```python
@timer
def loop(x):
    add=0
    for i in range(1,x):
        add=add+i
    return add
```

```python
class Cal():
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def add(self):
        return self.x+self.y
```

```python
obj1=Cal(10,20)
obj2=Cal(100,20)
```

```python
obj1.add()
obj2.add()
```

```python
class Dusri(Cal):
    
    def multiply(self):
        
        return self.x*self.y
```

```python
obj3=Dusri(2,3)
```

```python
obj3.multiply()
```

```python
obj3.add()
```

```python
class Third():
    def divide(self):
        return self.x/self.y
```

```python
class New(Cal,Third):
    pass
```

```python
obj4=New(2,10)
```

```python
obj4.add()
obj4.divide()
```

```python
class Grand():
    def grand(self):
        return f"I am grand Class"

class Parent(Grand):
    def parent(self):
        return f" I am Parent Class"

class Child(Parent):
    def child(self):
        return f" I am child class"

obj5=Child()

```

```python
obj5.grand()
obj5.parent()
obj5.child()
```

```python
#Encapsulation

class Cal():
    def __init__(self,x,y):
        self.x=x  #public
        self.__y=y #private

    def add(self):
        return self.x+self.y
```

```python
obj=Cal(1,2)
```

```python
obj.__y
```

```python
obj.y
```

```python
class Dog():
    def speak(self):
        return f" Bhoo Bhoo"

class Cat():
    def speak(self):
        return f" Meow"

obj=[Dog(),Cat()]

for animal in obj:
    print(animal.speak())
```

```python
class Cal1():
    def add(self,x,y):
        return x+y

class Cal2():
    def add(self,x,y,z):
        return x+y+z
```

```python
class Cal(Cal2,Cal1):
    pass
```

```python
obj=Cal()
```

```python
obj.add(1,2,3)
```
